# Fingerprints

```python
from rdkit.Chem import AllChem as Chem
from rdkit.Chem import Draw
from rdkit.Chem.Draw import IPythonConsole
import csv
```

```python
with open('Epidermal_growth_factor_receptor_erbB1_IC50.csv', 'r') as csvfile:
    reader = csv.DictReader(csvfile, delimiter=";")
    EGFR_ligands = [Chem.MolFromSmiles(m['Smiles']) for m in reader]

suppl = Chem.SDMolSupplier('../data/drugbank.sdf')
drugs = [m for m in suppl if m]
```

```python
EGFR_ligands_ecfp = [Chem.GetMorganFingerprintAsBitVect(m, 2, nBits=1024)
                     for m in EGFR_ligands]

drugs_ecfp = [Chem.GetMorganFingerprintAsBitVect(m, 2, nBits=1024) for m in drugs]
```

```python
EGFR_ligands_ecfp_sums = [0]*EGFR_ligands_ecfp[0].GetNumBits() # a list of zeros of a given length
for key in EGFR_ligands_ecfp:
    for onbit in key.GetOnBits():
        EGFR_ligands_ecfp_sums[onbit] += 1
EGFR_ligands_ecfp_sums
```

```python
drugs_ecfp_sums = [0]*drugs_ecfp[0].GetNumBits() # a list of zeros of a given length
for key in drugs_ecfp:
    for onbit in key.GetOnBits():
        drugs_ecfp_sums[onbit] += 1
drugs_ecfp_sums
```

```python
EGFR_ligands_ecfp_scaled = [x/len(EGFR_ligands_ecfp) for x in EGFR_ligands_ecfp_sums]
drugs_ecfp_scaled = [x/len(drugs_ecfp) for x in drugs_ecfp_sums]
```

```python
EGFR_drugbank_differences = [(i, a_b[0] - a_b[1])
                             for i, a_b in enumerate(zip(EGFR_ligands_ecfp_scaled, drugs_ecfp_scaled))]

EGFR_drugbank_differences.sort(key=lambda x: x[1])
EGFR_drugbank_differences[:3], EGFR_drugbank_differences[-3:]
```

The first three of the prevalent bits among EGFR ligands compared to DrugBank molecules are bits number 378, 428 and 491. The first three of the prevalent bits among DrugBank molecules compared to EGFR ligands are bits number 1, 389 and 147.

```python
def substructures_of_bit_ecfp4_1024(mol, bit):
    bitinfo = {}
    Chem.GetMorganFingerprintAsBitVect(mol, 2, nBits=1024, bitInfo=bitinfo)
    hit_atoms = [Chem.FindAtomEnvironmentOfRadiusN(mol, radius, atom)
                 for atom, radius in bitinfo.get(bit, [])
                 if radius > 0]
    submols = [Chem.PathToSubmol(mol, ha) for ha in hit_atoms]
    return submols
```

```python
substructure_smiles = {}
for ligand in EGFR_ligands:
    for submol in substructures_of_bit_ecfp4_1024(ligand, 378):
        subsmiles = Chem.MolToSmiles(submol)
        substructure_smiles[subsmiles] = substructure_smiles.get(subsmiles, 0) + 1
substructure_smiles
```

```python
substructure_smiles_db = {}
for drug in drugs:
    for submol in substructures_of_bit_ecfp4_1024(drug, 378):
        subsmiles = Chem.MolToSmiles(submol)
        substructure_smiles_db[subsmiles] = substructure_smiles_db.get(subsmiles, 0) + 1
substructure_smiles_db
```
